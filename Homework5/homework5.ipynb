```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# 下載停用詞庫
nltk.download('stopwords')
nltk.download('punkt')

# 載入停用詞
stop_words = set(stopwords.words('english'))

# 初始化 Porter 詞幹提取器
ps = PorterStemmer()

# 讀取論文文本
# 建立一個空的list，用於儲存所有txt檔案的data
all_data = []

all_data = []

# 逐一讀取每個txt檔案a
for i in range(1, 8):
    filename = f"0{i}.txt"  # 假設檔案名稱為 01.txt、02.txt、03.txt
    with open(filename, 'r', encoding='cp1252') as fh:
        tmp = fh.read()
        itemlist = tmp.split(',')
        
    all_data.append(itemlist)

# 將所有txt檔案的data疊加在一起
data = [item for sublist in all_data for item in sublist]


# 分詞、去除停用詞、詞幹提取
words = [ps.stem(word) for item in data for word in word_tokenize(item.lower()) if word.isalnum() and word not in stop_words]

```

```python
words
```

```python
import gensim
from gensim import corpora
from pprint import pprint
```

```python
# 創建詞袋
texts = [[word for word in document.lower().split()] for document in words]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# 訓練 LDA 模型
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, 
                                            num_topics=3, random_state=100, update_every=1, 
                                            chunksize=100, passes=10, alpha='auto', per_word_topics=True)
```

```python
# 輸出主題模型分析結果
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
