```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# 下載停用詞庫
nltk.download('stopwords')
nltk.download('punkt')

# 載入停用詞
stop_words = set(stopwords.words('english'))

# 初始化 Porter 詞幹提取器
ps = PorterStemmer()

# 讀取論文文本
# 建立一個空的list，用於儲存所有txt檔案的data
all_data = []

all_data = []

# 逐一讀取每個txt檔案a
for i in range(1, 8):
    filename = f"0{i}.txt"  # 假設檔案名稱為 01.txt、02.txt、03.txt
    with open(filename, 'r', encoding='cp1252') as fh:
        tmp = fh.read()
        itemlist = tmp.split(',')
        
    all_data.append(itemlist)

# 將所有txt檔案的data疊加在一起
data = [item for sublist in all_data for item in sublist]


# 分詞、去除停用詞、詞幹提取
words = [ps.stem(word) for item in data for word in word_tokenize(item.lower()) if word.isalnum() and word not in stop_words]

```

```python
words
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 將單詞列表轉為空格分隔的字符串
processed_text = ' '.join(words)

# 創建文字雲對象
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(processed_text)

# 顯示文字雲
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
